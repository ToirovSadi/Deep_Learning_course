#### Get Dataset


```python
from dataset import PH2Dataset
from torchvision import transforms

img_size = (256, 256)

root = 'PH2Dataset/PH2 Dataset images'
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor(),   
])

train_data = PH2Dataset(
    root=root,
    mode='train',
    transform=transform,
    download=True,
    target_dir='PH2Dataset'
)

val_data = PH2Dataset(
    root=root,
    mode='val',
    transform=transform,
    target_dir='PH2Dataset'
)

test_data = PH2Dataset(
    root=root,
    mode='test',
    transform=transform,
    target_dir='PH2Dataset'
)
```

#### Size of datasets

```python
print(f"train_dataset: {len(train_data)}")
print(f"val_dataset: {len(val_data)}")
print(f"test_dataset: {len(test_data)}")
```

### Lets visualize some of dataset images

```python
import matplotlib.pyplot as plt

num_images = 6
plt.figure(figsize=(18, 6))
for i in range(num_images):
    img = train_data[i][0]
    les = train_data[i][1]
    
    # (ch, w, h) -> (w, h, ch)
    img = img.permute(1, 2, 0).numpy()
    les = les.permute(1, 2, 0).numpy()
    
    plt.subplot(2, num_images, i + 1)
    plt.imshow(img)
    plt.title("Image")
    plt.axis("off")
    
    plt.subplot(2, num_images, i + num_images + 1)
    plt.imshow(les, cmap='gray')
    plt.title("Mask")
    plt.axis("off")
    
plt.show()
```
